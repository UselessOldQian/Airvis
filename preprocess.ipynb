```python
import pandas as pd
day = pd.read_table('day/20150123.txt', header=0,usecols=['Station_Id_d', 'Lat', 'Lon',
                                                   'WIN_D_Avg_2mi', 'WIN_S_Avg_2mi',
                                                         'Year','Mon','Day','Hour'],
                                                   encoding='gbk', sep='\s+')
day['date'] = day.apply(lambda x: convert(x),axis=1)
day['date'] = day['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d %H-%M-%S'))
day['date'] = day['date'].apply(lambda x: x.strftime('%Y%m%d%H'))
hour_set = set(day['date'])

for h in hour_set:
    day[day['date']==h].to_csv('result/'+h+'.txt',sep='\t',index=False)
```
